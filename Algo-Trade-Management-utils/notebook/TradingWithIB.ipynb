```python
# https://www.w3schools.com/python/trypython.asp?filename=demo_inheritance_child

class Person:
  def __init__(self, fname, lname):
    self.firstname = fname
    self.lastname = lname

  def printname(self):
    print(self.firstname, self.lastname)

class Animal:
  def say(self):
    print("ah!")

class Mamal:
  def mention(self):
    print("A person or man is a mamal!")
  def say(self):
    print("Mamal")

class Student(Person,Mamal,Animal):
  pass

x = Student("Mike", "Olsen")
x.printname()
x.say()  # if two super class has same method, the first one got inherinted!
x.mention()
```

```python
# fortune formula book, p198
# kelly vs markowitz Criterion
# both concur on diversification, especially on list 2

def arithmatic_mean(l):  # markowitz
    sum = 0
    for each in l:
        sum = sum + each
    res =  sum/len(l)
    return (res)

def geometric_mean(l):  # kelly
    prod = 1
    for each in l:
        prod = prod*each
    res = prod**(1/len(l)) 
    return (res)


list1 = [1,1,2,2,1,2]
list2 = [3,0,2,2,1,2]
list3 = [0.5,0.5,3,3,0.5,3]

list = list3
result = arithmatic_mean(list)
print(result)
result = geometric_mean(list)
print(result)
```
