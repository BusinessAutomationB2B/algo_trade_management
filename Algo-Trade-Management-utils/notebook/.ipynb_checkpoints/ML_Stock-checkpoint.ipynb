```python
# def trade signal marking function - prepare for supervised learning

def trade_signal(start,end, dataframe):
    for index, row in dataframe.iterrows():
        if str(row['Date']) == start:
            dataframe.at[index,'Signal'] = 'Buy'
        if str(row['Date']) > start and str(row['Date']) < end:
            dataframe.at[index,'Signal'] = 'Hold'
        if str(row['Date']) == end:
            df.at[index,'Signal'] = 'Sell' 


# df if passed in as original object which is modified

#mark trade 1
start = '2009-07-06'
end = '2009-10-18'
dataframe = df
trade_signal(start,end, dataframe)

#mark trade 2
start = '2010-06-29'
end = '2011-02-13'
dataframe = df
trade_signal(start,end, dataframe)       
        
#mark trade 3
start = '2012-05-11'
end = '2012-09-21'
dataframe = df
trade_signal(start,end, dataframe)   

#mark trade 4
start = '2012-11-11'
end = '2013-03-03'
dataframe = df
trade_signal(start,end, dataframe)

#mark trade 5
start = '2013-06-03'
end = '2014-06-04'
dataframe = df
trade_signal(start,end, dataframe)

#mark trade 6
start = '2015-01-11'
end = '2015-03-15'
dataframe = df
trade_signal(start,end, dataframe)

#mark trade 7
start = '2016-04-04'
end = '2016-08-15'
dataframe = df
trade_signal(start,end, dataframe)

#mark trade 8
start = '2016-10-30'
end = '2017-04-24'
dataframe = df
trade_signal(start,end, dataframe)

#mark trade 9
start = '2017-10-01'
end = '2018-09-02'
dataframe = df
trade_signal(start,end, dataframe)

#mark trade 10
start = '2018-12-23'
end = '2020-02-02'
dataframe = df
trade_signal(start,end, dataframe)

```
